```python
# Opo 1

import multiprocessing
import concurrent.futures
from datetime import datetime


def processamento(numero: int):
    print('# proc:', datetime.now().strftime("%H:%M:%S.%f"), '|', str(numero))


def main():
    workers = multiprocessing.Pool()._processes
    print('# workers:', workers)

    lista_numeros = list(range(0, 10000))
    tempo_inicio = datetime.now()
   
    with concurrent.futures.ThreadPoolExecutor() as executor:

        executor.map(processamento, lista_numeros)
            
    print('# Finalizado em:', (datetime.now() - tempo_inicio))


main()
```

```python
# Opo 2

import multiprocessing
import concurrent.futures
from datetime import datetime


def processamento(numero: int):
    print('# proc:', datetime.now().strftime("%H:%M:%S.%f"), '|', str(numero))


def main():
    workers = multiprocessing.Pool()._processes
    print('# workers:', workers)

    lista_numeros = list(range(0, 10000))
    tempo_inicio = datetime.now()
   
    executor = concurrent.futures.ThreadPoolExecutor()
    
    executor.map(processamento, lista_numeros)
    executor.shutdown(wait=True)
            
    print('# Finalizado em:', (datetime.now() - tempo_inicio))


main()
```
